```python
import pandas as pd
import numpy as np
import ast
import nltk
```

```python
movie1 = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movie1.columns
```

```python
credits= pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits.head()
```

```python
credits=credits.rename(columns={'movie_id': 'id'})
```

```python
movies=movie1.merge(credits,on='id')
```

```python
movies.columns
```

```python
movies=movies.drop(['budget','homepage','original_language','original_title','production_countries','release_date',
                    'revenue','runtime','spoken_languages', 'status','vote_average','vote_count','popularity','tagline'],axis=1)
```

```python
movies.columns
```

```python
movies=movies.drop('title_y',axis=1)
```

```python
movies=movies.rename(columns={'title_x' : 'title'})
```

```python
movies.columns
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert (obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[1].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[1].production_companies
```

```python
def convert1 (obj):
    L =[]
    count = 0
    for i in ast.literal_eval(obj):
        if count != 2:
            L.append(i['name'])
            count+=1 
        else:
            break
    return L
```

```python
movies['production_companies']=movies['production_companies'].apply(convert1)
```

```python
def convert2 (obj):
    L =[]
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count+=1 
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
def fetch_director(obj):
    L =[]  
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'genres', 'overview', 'production_companies', 'cast', 'crew','keywords']]
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['overview']=movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_companies']=movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies['tags']= movies['genres']+movies['overview']+ movies['production_companies']+movies['cast']+movies['crew']+movies['keywords']
```

```python
movies.head()
```

```python
new_df=movies[['id', 'title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features =4800,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
new_df.iloc[0].tags
```

```python
stem('action adventure fantasy sciencefiction in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. ingeniousfilmpartners twentiethcenturyfoxfilmcorporation samworthington zoesaldana sigourneyweaver jamescameron cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    # Find the index of the movie
    movie_index = new_df[new_df['title'] == movie].index[0]

    # Get similarity scores
    distances = similarity[movie_index]

    # Get top 7 most similar movies (excluding the first one which is the same movie)
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:7]

    # Print recommended movie indexes or titles
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
